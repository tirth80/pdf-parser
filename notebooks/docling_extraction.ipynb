````python
from docling.document_converter import DocumentConverter
import os
````

````python

# Can be local file path or URL
pdf_dir = "../data/raw/MSFT/10-K/PDFs/"

pdf_files = os.listdir(pdf_dir)
pdf_src = os.path.join(pdf_dir, pdf_files[0])
````

````python
pdf_src
````

````python
converter = DocumentConverter()
````

````python
result = converter.convert(pdf_src)    # returns a result with document

doc = result.document
````

````python
# Export to markdown
md = doc.export_to_markdown()
print(md)
````

````python
# Or JSON
js = doc.save_as_json("../data/parsed/output.json")
print(js)
````

#  Layout Extraction vs Docling Output Comparison


**Analysis Date**: September 26, 2025  
**Comparison Focus**: Structure, Content Quality, and Extraction Capabilities

---

##  **Executive Summary**

This document compares two PDF extraction approaches:
1. **Layout Extraction**: Custom LayoutLMv3-based approach with pdfplumber/OCR fallback
2. **Docling**: IBM's document intelligence platform with advanced AI models

### **Key Findings**
-  **Layout Extraction**: Better structured metadata, detailed page-level statistics
-  **Docling**: Superior semantic understanding, advanced content classification
- **Use Case**: Layout for analytics, Docling for content understanding

---

##  **Detailed Feature Comparison**

| Feature | Layout Extraction | Docling | Winner |
|---------|------------------|---------|---------|
| **Document Structure** |  Page-by-page layout |  Semantic hierarchy | 
| **Text Positioning** |  Precise coordinates (x0,y0,x1,y1) |  Limited positioning | 
| **Font Information** |  Name, size, color |  Not preserved | 
| **Table Extraction** |  Not structured |  Structured table objects | 
| **Figure Recognition** |  Not detected |  Figure objects with metadata | 
| **Content Classification** | Basic word-level |  Semantic content types | 
| **Processing Metadata** |  Detailed per-page stats |  Limited metadata | 
| **OCR Quality Assessment** |  Quality scores & flags |  Not provided |
| **Multi-column Detection** |  Column estimation |  Layout understanding |

##  **Use Case Recommendations**

### **Choose Layout Extraction When:**
-  **Precise positioning** is critical
- **Font analysis** is required
-  **Quality assessment** is needed
-  **Processing transparency** is important
-  **Custom post-processing** workflows
-  **Cost efficiency** is a priority

### **Choose Docling When:**
-  **Content understanding** is primary goal
-  **Document structure** preservation is key
- **Table extraction** is critical
-  **Figure handling** is required
-  **AI-powered analysis** is preferred
-  **Rapid prototyping** is needed
